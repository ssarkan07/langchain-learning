# OPEN AI Embedding Model


```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# Method 1: Using OpenAIEmbeddings from langchain_openai
from langchain_openai import OpenAIEmbeddings

simple_text= "Hello, I learning Langchain"
embeddings = OpenAIEmbeddings(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    check_embedding_ctx_length=False,
)
embedding_vectors = embeddings.embed_documents([simple_text])
print("Total Dimensions of the embedding vector:", len(embedding_vectors[0]))
print(embedding_vectors[0])
```

```python
#Method 2: Using OpenAI client directly
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
)

response = client.embeddings.create(
    model="openai/text-embedding-3-small",
    input=simple_text
)
print("Total Dimensions of the embedding vector:", len(response.data[0].embedding))
print(response.data[0].embedding[:5])
```

# Cosine similarity with OPENAI Embedding

```python
import numpy as np
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)
```

```python
sentences = [
    "Hello, I am learning Langchain",
    "Langchain is a powerful framework for building applications with LLMs",
    "I am excited to explore the capabilities of Langchain",
    "Langchain provides a seamless integration with various LLMs"
]

```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    api_key=os.getenv("OPEN_ROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    check_embedding_ctx_length=False,
)
```

```python
sentences_embeddings = embeddings.embed_documents(sentences)
print(sentences_embeddings[0])
```

```python
for i in range(len(sentences)):
    for j in range(i + 1, len(sentences)):
        similarity = cosine_similarity(sentences_embeddings[i], sentences_embeddings[j])
        print(f"Cosine similarity between sentence {i + 1} and sentence {j + 1}: {similarity:.4f}")
```

```python
# Semantic Search - Retrieving the most similar sentence to a query
query = "I want to learn about Langchain"
query_embedding = embeddings.embed_query(query)

# Find the most similar sentence to the query
for i, sentence_embedding in enumerate(sentences_embeddings):
    similarity = cosine_similarity(query_embedding, sentence_embedding)
    print(f"Cosine similarity between query and sentence {i + 1}: {similarity:.4f}")
```
