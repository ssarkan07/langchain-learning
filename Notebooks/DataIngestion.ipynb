## TextSplitting Technique

```python
from langchain_text_splitters import (
    RecursiveCharacterTextSplitter, CharacterTextSplitter, TokenTextSplitter
)
loader = TextLoader('Data/sample_text.txt')
text = loader.load()[0].page_content
```

```python
### Character Splitting

print("Character text splitter")
char_splitter = CharacterTextSplitter(
    separator=',', # split on ,
    chunk_size=250, # max chuck size in character
    chunk_overlap=20, # Overlap between chucks
    length_function=len #How to measure chucks size
)

char_chuck = char_splitter.split_text(text)
print("created", len(char_chuck))



```

```python
print(char_chuck[0])
print("-------------------")
print(char_chuck[1])
print("------------------")
print(char_chuck[2])
```

```python
# Recurssive Character Splitting (recommended)

print("Recurssive Character Text Splitter")
recur_splitter = RecursiveCharacterTextSplitter(
    separators=[',','.'], # split on ,
    chunk_size=250, # max chuck size in character
    chunk_overlap=20, # Overlap between chucks
    length_function=len #How to measure chucks size
)


recur_chuck = recur_splitter.split_text(text)
print("created", len(recur_chuck))
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
from langchain_core.documents import Document
import re
class SmartPDFProcessor:

    def __init__(self, chunk_size=1000, chunk_overlap=100):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap=chunk_overlap,
            separators=[" "],

        )

    def process_pdf(self, pdf_path: str) -> List[Document]:
        """process pdf with smart chunking and metadata enhancement"""

        ## load pdf
        loader = PyPDFLoader(pdf_path)
        pages = loader.load()

        ## process each page
        processed_chunks = []

        for page_num, page in enumerate(pages):
            cleaned_text = self._clean_text(page.page_content)

            # skip nearly empty pages
            if len(cleaned_text) < 50:
                continue


            # create the chunk with enhanced metadata
            chunks = self.text_splitter.create_documents(
                texts=[cleaned_text],
                metadatas=[{
                    **page.metadata,
                    "page": page_num+1,
                    "total_pages": len(pages),
                    "chunk_method": "smart_pdf_processor",
                    "char_count": len(cleaned_text)
                }]
            )
            processed_chunks.extend(chunks)
        return processed_chunks

    def _clean_text(self, text: str) -> str:
        text = text.replace("\f", " ")
        text = text.replace("\xa0", " ")

        # Fix hyphenated words across lines
        text = re.sub(r"(\w)-\s*\n\s*(\w)", r"\1\2", text)

        # Remove standalone page numbers
        text = re.sub(r"\bPage\s+\d+\b", "", text, flags=re.IGNORECASE)

        # Replace newlines with spaces
        text = text.replace("\n", " ")

        # Normalize whitespace
        text = re.sub(r"\s+", " ", text)

        return text.strip()



```

```python
preprocess = SmartPDFProcessor()

processed_chunk = preprocess.process_pdf('Data/Attention Mechanism.pdf')
print(len(processed_chunk))
```
